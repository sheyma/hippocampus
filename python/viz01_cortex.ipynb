# LSUB

```python
fname = '../data/tout_group/cortex709_LSUB.h5'

h = h5py.File((fname), 'r')
data = np.array(h['LSUB'])
h.close()
data = data.reshape(data.shape[0])

data.shape
```

```python
data.min(), data.max()
```

```python
# load surfaces
surf_lh, surf_rh = load_conte69()

# load labeling file
labeling_file = '../data/tout_group/glasser.csv' 
labeling = genfromtxt(labeling_file)

mask = labeling != 0 # this takes the mask of the left hem only..

mymask = np.ones((64984,)).astype('bool')
mymask[np.where(mask == False)[0]] = False
mymask[np.where(mask == False)[0] + 32492] = False

data_labeled = map_to_labels(data, labeling, mask=mymask) 
data_labeled[np.where(mymask == False)] = np.nan

data_labeled.shape
```

```python
plot_hemispheres(surf_lh, surf_rh, 
                 array_name = data_labeled,
                 nan_color = (0.9,0.9,0.9,1),
                 size = (1200, 200),
                 cmap = 'inferno', 
                 color_bar = True,
                 color_range = (0, 0.45),
                 interactive = False,
                 zoom = 1.3,
                 embed_nb = True,
                 #screenshot = True,
                 #filename = '../figures/cortex_LSUB.png'
                 )
```

# Tvals

```python
import sys
sys.path.append("/data/p_02323/BrainStat/surfstat/") 
sys.path.append("/data/p_02323/BrainStat/surfstat/python") 
from SurfStatSmooth import py_SurfStatSmooth
from brainspace.mesh import mesh_elements
from brainspace.datasets import load_conte69
```

```python
surf_lh, surf_rh = load_conte69()

# write surface coordinates and triangles in a dictionary
lh_coord = np.array(mesh_elements.get_points(surf_lh)).T
rh_coord = np.array(mesh_elements.get_points(surf_rh)).T
lh_tri = np.array(mesh_elements.get_cells(surf_lh))
rh_tri = np.array(mesh_elements.get_cells(surf_rh))

D = {}
D['coord'] = np.concatenate((lh_coord, rh_coord), axis=1)         # (3, 64984)     
D['tri'] = np.concatenate((lh_tri, rh_tri + lh_coord.shape[1]))   # (129960, 3)

```

```python
# load labeling file
labeling_file = '../data/tout_group/glasser.csv' 
labeling = genfromtxt(labeling_file)

mask = labeling != 0 # this takes the mask of the left hem only..

mymask = np.ones((64984,)).astype('bool')
mymask[np.where(mask == False)[0]] = False
mymask[np.where(mask == False)[0] + 32492] = False
```

## cortex LSUB

```python
fname = '../data/tout_group/Tvals_cortex709_LSUB.h5'

h = h5py.File((fname), 'r')
Tdata = np.array(h['data'])
h.close()

Tdata = Tdata.reshape(Tdata.shape[1],)
print(Tdata.shape)


Tsmoo = py_SurfStatSmooth(Tdata.reshape(len(Tdata),1).T, D, 3)
Tsmoo = Tsmoo.reshape(Tsmoo.shape[1],)
print(Tsmoo.shape)
```

```python
Tsmoo[Tsmoo<20] = np.nan
Tsmoo[np.where(mymask == False)] = np.nan
```

```python
plot_hemispheres(surf_lh, surf_rh, 
                 array_name = Tsmoo,
                 nan_color = (0.9,0.9,0.9,1),
                 size = (1200, 200),
                 cmap = plot_funcs.hotcolors, 
                 color_bar = True,
                 color_range = (20, 35),
                 interactive = False,
                 zoom = 1.3,
                 embed_nb = True,
                 #screenshot = True,
                 #filename = '../figures/Tvals_cortex_LSUB.png'
                 )
```
